## Code

### Load Libraries

First, we need to import the necessary libraries for our linear regression model. We will use Numpy, as well as scikit-learn library and matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model
import sklearn.model_selection
import sklearn.metrics
```

##### Load Data

In this case, the data used for the linear regression model will be random. The dataset will be generated using the "random" function from Numpy.

```python
# Generation of Random Data Using Numpy

np.random.seed(0)
study_hours = np.random.rand(100, 1) * 10
grades = 2 * study_hours + np.random.randn(100, 1) * 2
```

#### Data Preprocessing

We preprocess our generated data according to the labels of our linear regression model. We also stablish our test data and test labels.

```python

# Labels from generated data
X = np.array(study_hours)
Y = np.array(grades)

# Split the train and testing data into sets (80% / 20%)
train_data, test_data, train_labels, test_labels = sklearn.model_selection.train_test_split(X, Y)
```

#### Model Selection and Training

We select the linear regression model and we train the model with the preprocess data.

```python
# Linear Regression Model
model = sklearn.linear_model.LinearRegression()

# Train Model
model.fit(X, Y)
```

#### Model Testing

We test the model using the model.predict()  and obtain the predicted data generated by our linear regression model.

```python
predict_data = model.predict(test_data)
print(predict_data)
```

With the use of the matplotlib library, we can visualize our linear regression model.

```python
plt.scatter(X, Y, color='steelblue')
plt.xlabel("Study Hours")
plt.ylabel("Grades")
plt.title("Linear Regression for Grades Prediction")
plt.ylim(0, 20)
plt.plot(X, model.predict(X), color='red')
```

Finally, we calculate the mean squared error.

```python
mean_squared_error = sklearn.metrics.mean_squared_error(test_labels, predict_data)
print("Mean Squared Error:", mean_squared_error)
```
